```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import exp, cos


t = np.arange(0, 200, 0.01)
y1 = [1-0.29*exp(-10*tp)-1.19*exp(-2*tp)*cos(4.53*tp - 53.3) for tp in t]
y2 = [1-0.29*exp(-10000*tp)-1.19*exp(-2*tp)*cos(4.53*tp - 53.3) for tp in t]
y3 = [1-0.29*exp(-1000*tp)-1.19*exp(-2*tp)*cos(4.53*tp - 53.3) for tp in t]
y4 = [1 for tp in t]

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
plt.title("Different exponentials", color="white")
ax.scatter(t, y1, color="red")
ax.scatter(t, y2, color="green")
ax.scatter(t, y3, color="blue")
ax.plot(t, y1, 'r')
ax.plot(t, y2, 'g')
ax.plot(t, y3, 'b')
ax.plot(t, y4, '--g')
ax.set_xlabel("Time")
ax.set_ylabel("Response")
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
ax.spines['bottom'].set_color('white')
ax.spines['top'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['right'].set_color('white')
ax.set_ylim((0, 2.5))
ax.set_xlim((0,3))
plt.grid(True)
```
